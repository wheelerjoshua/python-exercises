## 3.

```python
import json

profiles = json.load(open('profiles.json'))
profiles
```

### Number of users

```python
total_users = len(profiles)
total_users
```

### Number of active users

```python
total_active = 0
for user in profiles:
    if user['isActive'] == True:
        total_active += 1
total_active
```

### Number of inactive users

```python
total_inactive = 0
for user in profiles:
    if user['isActive'] == False:
            total_inactive += 1
total_inactive
```

### Total of balances for all users

```python
total_balance = 0
for user in profiles:
    balance = user['balance']
    while balance[0].isnumeric() == False:
        balance = balance[1:]
    balance = functions_exercises.handle_commas(balance)
    total_balance += balance
total_balance
    
```

### Average balance per user

```python
average_balance = total_balance / total_users
average_balance
round(average_balance, 2)
```

### User with lowest balance

```python
lowest_balance = 1000000
for user in profiles:
    test_balance = user['balance']
    while test_balance[0].isnumeric() == False:
        test_balance = test_balance[1:]
    test_balance = functions_exercises.handle_commas(test_balance)
    if test_balance < lowest_balance:
        lowest_balance = test_balance
        lowest_balance_user = user['name']

lowest_balance_user
```

### User with Highest Balance

```python
highest_balance = 0
for user in profiles:
    test_balance = user['balance']
    while test_balance[0].isnumeric() == False:
        test_balance = test_balance[1:]
    test_balance = functions_exercises.handle_commas(test_balance)
    if test_balance > highest_balance:
        highest_balance = test_balance
        highest_balance_user = user['name']
        
highest_balance_user
```

### Most common favorite fruit

```python
favorite_fruits = []
for user in profiles:
    favorite_fruits.append(user['favoriteFruit'])

max_fruit = 0
for fruit in favorite_fruits:
    fruit_count = favorite_fruits.count(fruit)
    if fruit_count > max_fruit:
        max_fruit = fruit_count
        most_common = fruit


most_common
```

```python
max(favorite_fruits, key=favorite_fruits.count)
```

```python
least_favorite = 1000
for fruit in favorite_fruits:
    fruit_count = favorite_fruits.count(fruit)
    if fruit_count < least_favorite:
        least_favorite = fruit_count
        least_common = fruit
        
least_common
```

```python
min(favorite_fruits, key=favorite_fruits.count)
```

### Total number of unread messages

```python
total_user_unread = 0
for user in profiles:
    greeting = user['greeting']
    total_user_unread_str = ''
    for char in greeting:
        if char.isnumeric():
            total_user_unread_str += char
    total_user_unread += int(total_user_unread_str)
    
total_user_unread
```
